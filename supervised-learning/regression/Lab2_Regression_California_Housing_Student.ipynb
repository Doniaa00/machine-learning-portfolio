# Lab 2 — Regression (California Housing)

**Objective:** Predict median house value for California census block groups.

**Dataset:** California Housing — 20,640 samples, 8 features (MedInc, HouseAge, AveRooms, etc.), target = median house value (in 100,000 USD).

## Step 1: Imports and load data

Import the required libraries and load the California Housing dataset. Store features in `X` and target in `y`.

```python
# Step 1: Imports and load data
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Load California Housing data
data = fetch_california_housing()
X = data.data
y = data.target

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Feature names:", data.feature_names)
```

## Step 2: Train–test split

Split `X` and `y` into training and test sets with `test_size=0.3`.

```python
# Step 2: Train–test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Train size:", len(y_train))
print("Test size:", len(y_test))
```

## Step 3: Train the model

Create a `LinearRegression` model, fit it on the training data, and predict on the test set.

```python
# Step 3: Train the model
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("Model training completed")
```

## Step 4: Evaluate

Print the Mean Squared Error between `y_test` and `pred`.

```python
# Step 4: Evaluate
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)

print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
```
