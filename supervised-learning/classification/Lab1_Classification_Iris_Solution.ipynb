# Lab 1 — Classification (Iris Dataset) — SOLUTION

**Objective:** Build a classifier that predicts iris species from flower measurements.

**Dataset:** Iris — 150 samples, 4 features, 3 classes (Setosa, Versicolor, Virginica).

## Step 1: Imports and load data

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

data = load_iris()
X = data.data
y = data.target

print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Step 2: Train–test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print("Train size:", len(y_train))
print("Test size:", len(y_test))
```

## Step 3: Train the model

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

## Step 4: Evaluate

```python
print(classification_report(y_test, pred))
```
